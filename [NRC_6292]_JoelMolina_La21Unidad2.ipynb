 <h3> <font color='orange'>VALORES NULOS</font></h3> 

```python
''' Verificamos los valores nulos de las columnas de nuestro DataSet'''
valoresnulos_Data = df_Datos.isna().sum()
valoresnulos_Data
```

Como podemos visualizar, tenemos que hay una cantidad extensa de datos nulos en la columna Range. Por lo que, procedemos a eliminar dicha columna.

Como podemos verificar los registros de la columna Disposition hay dos valores q no concuerdan con los demás datos, estos valores son Not recorded y 22. Por lo que, realizamos el cambio de estos dos registros por un valor concordante al conjunto de datos. Al registro Not recorded le cambiamos el valor por las siglas NR y el 22 por el valor de la moda.

```python
'''Calculamos la moda de la columna Disposition '''
moda= df_DatosLimpios['Disposition'].mode()
'''Imprimimos la moda'''
moda
```

```python
'''Se realizan cambios en la columna Disposition'''
df_DatosLimpios['Disposition']= df_DatosLimpios['Disposition'].replace('22', moda[0])
df_DatosLimpios['Disposition']= df_DatosLimpios['Disposition'].replace('Not recorded', 'NR')
```

Como podemos visualizar en la columna City hay valores no coherentes como el registro San Franciso que está escrito todo en mayúsculas y otros registros escrito de la siguiente manera: "San Francisco". Por otro lado, tambien podemos visualizar que esxisten registros con valores S. Por lo que, realizamos el cambio de los datos de San Francisco en mayúscula por el siguiente formato:"San Francisco". Además, los registros con el valor de "S" de igual manera se cambian a "San Francisco". 

```python
 '''Se realizan cambios en la columna Disposition'''
df_DatosLimpios['City']= df_DatosLimpios['City'].replace('SAN FRANCISCO', 'San Francisco')
df_DatosLimpios['City']= df_DatosLimpios['City'].replace(' S', 'San Francisco')
```

Por lo que, se realiza el cambio de su valor, por su valor correcto. Además, el registro con el valor 1 lo cambiamos por la moda.
